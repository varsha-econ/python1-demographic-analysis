```python
# Section A

## Question 1:
c1 = '`sex`'
c2 ='`age`'
c3 ='`survived`'
s1 = 'female'

df_c1 = df.query('{0} == @s1 & ({1} < 18 & {2} ==1 )'.format(c1,c2,c3))
df_c1

a = df_c1['name']
print('all passengers who are female,under age 18, and survived are:',a)


```

```python
## Question 2:

average_fare = df['fare'].mean()
print(average_fare)

# the average fare is 34.694514005602244
c1 = '`fare`'
c2 ='`pclass`'

df_c2 = df.query(('{0} > 34 & {1} == 1').format(c1,c2))
df_c2
df_c2['name']
print('the names of passengers who paid more than average fare and travelled in 1st class are:',df_c2['name'])
```

```python
# Section B

## Question 3:
df_by_pclass = (df.groupby('pclass').mean(numeric_only=True))
print(df_by_pclass)

## Question 4:

df_by_sex_mean = (df.groupby('sex').mean(numeric_only=True))
print('the mean values are :',df_by_sex_mean)

df_by_sex_median = (df.groupby('sex').median(numeric_only=True))
print('the median values are',df_by_sex_median)
```
